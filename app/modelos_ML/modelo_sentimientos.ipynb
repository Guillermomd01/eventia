```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import joblib
```

```python
df = pd.read_csv('dataset_2017_test.csv')
```

```python
#filtramos y estilizamos el dataset
df.columns=['id','texto','emocion']
df = df.loc[df['emocion']!='NONE']
df['texto'] = df['texto'].str.lower()
```

```python
import re

df['texto'] = df['texto'].str.replace(r'@\w+', '', regex=True)

df
```

```python
#creamos las variables X e y
X = df['texto']
y= df['emocion']
```

```python
#creamos el vectorizador vectorizamos el texto
tfidf = TfidfVectorizer(max_features=4000)
```

```python
X_vectores = tfidf.fit_transform(X)
```

```python
#establecemos los porcentajes de entramiento y test
X_train,X_test,y_train,y_test = train_test_split(X_vectores,y,test_size=0.2,random_state=42)
```

```python
#convertimos a array para que LogisticRegression lo pueda soportar
X_train = X_train.toarray()
X_test = X_test.toarray()
```

```python
#elegimos el modelo LogisticRegression

mlr = LogisticRegression()
mlr.fit(X_train, y_train)
```

```python
#Calculamos la precision del modelo entrenado

prediciones = mlr.predict(X_test)
accuracy = accuracy_score(y_test, prediciones)
print(f'Precisión: {accuracy:.2f}')
```

```python
#reporte con distitos datos del modelo

from sklearn.metrics import classification_report
print(classification_report(y_test,prediciones))
```

```python
#pobamos otro modelo para comparar precisiones

from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

model_multi = MultinomialNB()
model_multi.fit(X_train, y_train)
predicion = model_multi.predict(X_test)

precision = accuracy_score(y_test,predicion)
print(precision)
```

```python
#pobamos otro modelo para comparar precisiones


from sklearn.svm import LinearSVC
svc = LinearSVC()
svc.fit(X_train, y_train)
pred = svc.predict(X_test)

preci=accuracy_score(y_test,pred)
print(preci)
```

```python
#importamos el mejor modelo: LogisticRegresion
joblib.dump(mlr,'modelo_entrenado_predicion.pkl')
```

```python
joblib.dump(tfidf,'vectorizador_entrenado.pkl')
```
